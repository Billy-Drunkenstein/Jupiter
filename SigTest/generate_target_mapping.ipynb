```python
import json
import pandas as pd
from pathlib import Path
```

```python
target_feather_path = Path("/gpfs/hddfs/shared/tzheng_ryin/cnfut/cnfut_snap_y_feather")
feather_files = list(target_feather_path.glob("*.feather"))
sample_file = feather_files[0]
```

```python
df = pd.read_feather(sample_file)
target_cols = [c for c in df.columns if c.startswith("y")]
target_mapping = {col: f"Y{i}" for i, col in enumerate(target_cols)}
```

```python
output_path = Path("target_mapping.json")
with open(output_path, "w") as f:
    json.dump(target_mapping, f, indent = 2)
```
